```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style='darkgrid')
from hidden import dynamics
from hidden import infer
```

```python
hmm = dynamics.HMM(2, 2)
hmm.init_uniform_cycle(0.3, 0.2)
mat_perturb = np.array([
    [-0.05, 0.04],
    [0.05, -0.04]
])

A_sample = hmm.A + mat_perturb
B_sample = hmm.B + mat_perturb

hmm.run_dynamics(200)
obs_ts = hmm.get_obs_ts()
state_ts = hmm.get_state_ts()

analyzer = infer.MarkovInfer(2, 2)
analyzer.bayesian_smooth(obs_ts, A_sample, B_sample)
```

```python
fig, ax = plt.subplots(1, 1)

ax.plot(obs_ts[10: 70], 'o', color='r')
ax.plot(state_ts[10: 70], 'o', color='b')
ax.plot(analyzer.bayes_smooth[10: 70, 1], 'o', color='g')
ax.plot(analyzer.bayes_smooth[10: 70, 1], color='g')
```
