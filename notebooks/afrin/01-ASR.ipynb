```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt


mu, sigma = 0, 1 
sample_size = 1000
sample_data = np.random.normal(mu, sigma, sample_size)

# empirical probability density function
bins = 30
density, bins, _ = plt.hist(sample_data, bins=bins, density=True)

# analytical steady-state distribution
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
analytical_density = stats.norm.pdf(x, mu, sigma)

# Plot the two distributions for comparison
plt.plot(x, analytical_density, label='Analytical')
plt.plot(bins[:-1], density, label='Empirical')
plt.legend()
plt.show()
```
